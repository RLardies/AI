## 1. Construcción de clasificadores en bases de datos sintéticas

```python
# Esta primera parte está adaptada de:
# http://scikit-learn.org/stable/auto_examples/classification/plot_classifier_comparison.html
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons, make_circles
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier

%matplotlib inline
```

```python
# data preparation
simple = make_blobs(n_samples=500, n_features=2,
                    centers=[[0,0], [2.5,2.5]],
                    random_state=1)
X,y = make_blobs(n_samples=500, n_features=2,
                 centers=[[0,0], [5,1]],
                 random_state=1)
X = X*np.matrix([[1,-2],[-20,10]])
linearly_separable=(X,y)

datasets = [simple,
            linearly_separable,
            make_moons(noise=0.1, random_state=0, n_samples=500),
            make_circles(noise=0.1, factor=0.5, random_state=1,
                         n_samples=500)
           ]
```

```python
classifiers = [("Naive Bayes", GaussianNB()),
               ("Nearest Neighbors",
                KNeighborsClassifier(n_neighbors=5)), # número de vecinos
               ("Decision Tree",
                DecisionTreeClassifier(criterion='entropy',
                                       max_depth=7)), # profundidad máxima del árbol
               ("Logistic Regression",
                LogisticRegression(C=1e10,solver='lbfgs')), # C: cuanto más alto menos regularización 
               ("Neural Network",
                MLPClassifier(hidden_layer_sizes=(10,10,10,),
                              max_iter=1000,
                              alpha=0))]

from p4_IA_aux import plot_classifiers

plot_classifiers(classifiers, datasets)
```

```python
# Listado de los parámetros de los diferentes clasificadores
for name, clf in classifiers:
    print(clf)
```

**Prueba a cambiar los siguientes parámetros y observa las consecuencias en la frontera de clasificación construida:**

* Número de vecinos en k-nn. ¿Por qué siempre debe ser impar cuando hay dos clases?
* Profundidad máxima de los árboles de decisión.
* Número de neuronas en la red neuronal y máximo número de épocas de entrenamiento. **Nota:** (50,) indica una única capa oculta con 50 neuronas. (50,10,) indica dos capas ocultas con 50 y 10 neuronas respectivamente. (50,10,20,) indica tres capas ocultas con 50, 10 y 20 neuronas respectivamente, etc.

* Para **k-nn**, al ir aumentando el numero de vecinos, observamos que las fronteras se mantienen muy similares, y van apareciendo entorno a estas puntos con menor probabilidad a acertar. Esto se debe a que al tener más vecinos y estar en la frontera, estos tendrán vecinos de ambas clasesn por lo que disminuirá la probabilidad de acierto en la predicción. El **numero de vecinos** en k-nn ha de ser siempre impar para evitar empates en las decisiones entre dos clases, ya que siempre habrá mayoría para alguna de las dos.


* Para **Decision tree**, conforme vamos aumentando la profundidad del arbol, aumentamos la precisión considerablemente, hasta que llegamos a profundidad 7 (desde 0,91 a 0,98). A partir de esta profundidad las variaciones en los datos obtenidos son mínimas. Esto se produce porque la librería sklearn, para evitar el problema de "overfitting" añade por defecto los parametros ***min_samples_leaf=1*** y ***min_samples_split=2*** . esto quiere decir que los datos se seguirán dividiendo solo si hay al menos dos muestras en esa región. Si no los hay el arbol ignorará esa hoja y no seguirá dividiendo.

* Para las redes neuronales, examinamos primero el caso de **una única capa oculta**. Observamos que para un número bajo de neuronas, no converge en todos los casos para 1000 iteraciones, aunque a partir de 20 neuronas, los resultados son bastante buenos (0.94, 0.99, 0.95, 0.99). A partir de unas 30 neuronas converge para todos los casos con una precisión de (0.95, 0.99, 0.98, 0.99). Aumentar más neuronas no cambia el resultado prácticamente, incluso lo empeora (para 40 neuronas (0.94, 0.99, 0.98, 0.99)) debido al problema de "overfitting".

* Para el caso de **dos capas ocultas**, tomando valores pequeños como 8:6 (8 neuronas la primera capa, 6 la segunda) converge para 1000 iteraciones con unos valores de precisión no muy buenos (0.94, 0.99, 0.86, 0.97). Aumentando las neuronas de la segunda capa a 8:8 obervamos un aumento muy considerable de la precisión (0.95, 0.99, 0.99, 0.99). Aumentando las neuronas de ambas capas la mejora de la precisión es mínima. Por ejemplo, para 20:20 obtenemos (0.94, 0.99, 1.00, 0.99). Observamos que al añadir capas, necesitamos menos neuronas para obtener resultados similares. Necesitabamos 30 neuronasen una capa para obtener una precisión aceptable, mientras que con dos capas hemos necesitado únicamente 16 neuronas.

* Para el caso de **tres capas ocultas**, el valor más pequeño para el que convergen que hemos encontrado es 6:4:4. Los resultados obtenidos son (0.95, 0.99, 0.99, 0.98). Al aumentar las neuronas, la mejora en la precisión vuelve a ser prácticamente nula. Por ejemplo, para 10:10:10 obtenemos (0.95, 0.99, 1.00, 0.98). Las neuronas necesarias para obtener este resultado son prácticamente las mismas que con dos capas pero al añadir una capa más, la complejidad aumenta sustancialmente. Por esto, llegamos a la conclusión, que lo ideal para este problema, sería usar dos capas de aproximadamente 8 neuronas cada una.
